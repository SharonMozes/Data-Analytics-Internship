```python
x=np.array([11,45,67,7,33,56,34,22])
```

```python
def normalize(a):
    norm=(a-a.min())/(a.max()-a.min())
    return norm
```

```python
normalize(x)
```

```python
nx=normalize(x)
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.title('Actual Data')
plt.plot(x);
plt.subplot(1,2,2)
plt.title('Normalize Data')
plt.plot(nx);
```

```python
df=pd.read_csv("Social_Media_Ads.csv")
```

```python
x=df[['Age','EstimatedSalary']]
y=df['Purchased']
```

```python
plt.xlabel('age')
plt.ylabel('salary')
plt.grid()
plt.scatter(x['Age'],x['EstimatedSalary'],c=y)
```

```python
x.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
x_scaled=scaler.fit_transform(x)
```

```python
type(x_scaled)
```

```python
x_new=pd.DataFrame(x_scaled,columns =x.columns)
```

```python
x_new
```

```python
x_new.describe()
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.title('Actual Data')
plt.xlabel('age')
plt.ylabel('salary')
plt.grid()
plt.scatter(x['Age'],x['EstimatedSalary'],c=y)

plt.subplot(1,2,2)
plt.title('Normalize Data')
plt.xlabel('age')
plt.ylabel('salary')
plt.grid()
plt.scatter(x_new['Age'],x['EstimatedSalary'],c=y)

```
